```
# Padronizar nomes de disciplinas
def limpar_disciplina(texto):
    if not isinstance(texto, str):
        return ""
    texto = re.sub(r'\s[A-Z][0-9]-.*', '', texto.upper().strip())
    texto = re.sub(r'\s\(S[AB]\)$', '', texto)
    return texto.strip()

df_matriculas['Disciplina_Limpa'] = df_matriculas['TURMA'].apply(limpar_disciplina)
print(f"{len(df_matriculas['Disciplina_Limpa'].unique())} disciplinas únicas")
```

### 4. Criação do Banco de Dados SQLite

```
import sqlite3

# Normalização nomes das colunas - Matrículas
df_matriculas_clean = df_matriculas.copy()
new_columns = []
col_counts = {}

for col in df_matriculas_clean.columns:
    clean_col = str(col).strip().replace(' ', '_')
    clean_col = ''.join(c if c.isalnum() or c == '_' else '' for c in clean_col)

    if clean_col.lower() in col_counts:
        col_counts[clean_col.lower()] += 1
        clean_col = f"{clean_col}_{col_counts[clean_col.lower()]}"
    else:
        col_counts[clean_col.lower()] = 0

    new_columns.append(clean_col)

df_matriculas_clean.columns = new_columns
print(f"{len(new_columns)} colunas processadas na tabela de matrículas")
print(f"\nColunas na matricula: {list(df_matriculas_clean.columns)}")
```

```
df_catalogo_clean = df_catalogo.copy()
new_columns_catalogo = []
col_counts_catalogo = {}

for col in df_catalogo_clean.columns:
    clean_col = str(col).strip().replace(' ', '_')
    clean_col = ''.join(c if c.isalnum() or c == '_' else '' for c in clean_col)

    if clean_col.lower() in col_counts_catalogo:
        col_counts_catalogo[clean_col.lower()] += 1
        clean_col = f"{clean_col}_{col_counts_catalogo[clean_col.lower()]}"
    else:
        col_counts_catalogo[clean_col.lower()] = 0

    new_columns_catalogo.append(clean_col)

df_catalogo_clean.columns = new_columns_catalogo
print(f"{len(new_columns_catalogo)} colunas processadas na tabela de catálogo")
print(f"\nColunas no catálogo: {list(df_catalogo_clean.columns)}")
```

```
# Criação do banco SQLite com duas tabelas
db_path = "matriculas_ufabc_v2.db"
conn = sqlite3.connect(db_path)

# Inserir tabela de matrículas
df_matriculas_clean.to_sql('matriculas', conn, if_exists='replace', index=False)

# Inserir tabela de catálogo
df_catalogo_clean.to_sql('catalogo', conn, if_exists='replace', index=False)

# Verificar criação
cursor = conn.cursor()
cursor.execute("SELECT COUNT(*) FROM matriculas")
total_matriculas = cursor.fetchone()[0]

cursor.execute("SELECT COUNT(*) FROM catalogo")
total_catalogo = cursor.fetchone()[0]

conn.close()

print(f"Banco criado com 2 tabelas:")
print(f"  - matriculas: {total_matriculas} registros")
print(f"  - catalogo: {total_catalogo} registros")
```

```
# Verificar se SIGLA existe em ambas as tabelas
conn = sqlite3.connect(db_path)

# Colunas da tabela matriculas
cursor = conn.cursor()
cursor.execute("PRAGMA table_info(matriculas)")
cols_matriculas = [row[1] for row in cursor.fetchall()]

# Colunas da tabela catalogo
cursor.execute("PRAGMA table_info(catalogo)")
cols_catalogo = [row[1] for row in cursor.fetchall()]

print("Verificando coluna SIGLA:")
print(f"  - Matrículas: {'SIGLA' in cols_matriculas}")
print(f"  - Catálogo: {'SIGLA' in cols_catalogo}")

# Teste de JOIN
if 'SIGLA' in cols_matriculas and 'SIGLA' in cols_catalogo:
    query = """
    SELECT COUNT(*)
    FROM matriculas m
    INNER JOIN catalogo c ON m.SIGLA = c.SIGLA
    """
    cursor.execute(query)
    total_join = cursor.fetchone()[0]
    print(f"\n✓ JOIN possível: {total_join} registros relacionados")
else:
    print("\n⚠ SIGLA não encontrada em uma ou ambas as tabelas")
    print(f"\nColunas em matriculas: {cols_matriculas[:5]}...")
    print(f"Colunas em catalogo: {cols_catalogo[:5]}...")

conn.close()
```

### 5. Configuração do SQL Agent

```
# Conectar ao banco via LangChain
from langchain_community.utilities import SQLDatabase

db = SQLDatabase.from_uri(f"sqlite:///{db_path}")
print(f"Conectado ao banco. Tabelas: {db.get_usable_table_names()}")
```

```
# Configurar LLM
from langchain_groq import ChatGroq

llm = ChatGroq(temperature=0.1, model_name="llama-3.3-70b-versatile")

print("LLM configurado: Llama 3.3 (70B)")
```

```
# Criar toolkit de ferramentas SQL
from langchain_community.agent_toolkits import SQLDatabaseToolkit

toolkit = SQLDatabaseToolkit(db=db, llm=llm)
print(f"Toolkit criado com {len(toolkit.get_tools())} ferramentas")
```

```
# Criar SQL Agent
from langchain_community.agent_toolkits import create_sql_agent

prefix = """
Você é um agente SQL para consultas sobre matrículas da UFABC.

Processo:
1. Liste as tabelas (sql_db_list_tables)
2. Consulte o schema (sql_db_schema)
3. Gere a query SQL
4. Valide a query (sql_db_query_checker)
5. Execute (sql_db_query)
6. Interprete em português a resposta
7. Responda usando o formato `Final Answer: sua resposta aqui`.

Informações da tabela 'matriculas':
- Disciplina_Limpa: nome padronizado da disciplina (PK)
- DOCENTE_TEORIA: professor (pode ser NULL)
- TEORIA/PRÁTICA: horários de disciplinas e aulas
- TURNO: Matutino/Noturno/Vespertino
- CAMPUS: SA/SB

Tabela 'catalogo':
- SIGLA: código da disciplina (CHAVE PRIMÁRIA)
- DISCIPLINA: nome da disciplina
- RECOMENDAÇÃO: matérias recomendadas a esta disciplina
- OBJETIVOS: objetivos da disciplina
- EMENTA: ementa da disciplina

Diretrizes:
- SEMPRE use UPPER() ao comparar strings ou converta o termo de busca para maiúsculas
- Use LIKE com % para buscas
- Use DISTINCT para evitar duplicatas
- Limite a 10 resultados
- Filtre DOCENTE_TEORIA IS NOT NULL para professores
- Responda em português
"""

agente = create_sql_agent(
    llm=llm,
    toolkit=toolkit,
    agent_type="zero-shot-react-description",
    prefix=prefix,
    verbose=True,
    max_iterations=15,
    handle_parsing_errors=True
)

print("SQL Agent criado")
```

```
# @title 3. ETL e Criação da Base Vetorial (Enriquecido)
import pandas as pd
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document

# 1. Carregamento dos Dados
SHEET_ID = "1fAKm4gI91rirQq-UqxQmSTiOxG9-YnwyKTJI1mFOKF8"
CSV_URL = f"https://docs.google.com/spreadsheets/d/{SHEET_ID}/export?format=csv"

print(f"Baixando dados de: {CSV_URL} ...")

try:
    df = pd.read_csv(CSV_URL)
    df.columns = df.columns.str.strip()
    df.fillna("Não informado", inplace=True)

    print(f"Dados carregados. Linhas: {len(df)}")
    print(f"Colunas: {df.columns.tolist()}")

    documents = []
    for index, row in df.iterrows():
        nome = row.get('Nome', 'Professor Desconhecido')
        email = row.get('Email ', row.get('Email', 'Email não informado'))
        linha_pesquisa = row.get('Linha de Pesquisa', 'Área não informada')
        ic = row.get('Edital', 'Não informado')
        pdpd = row.get('Unnamed: 4', 'Não informado')


        page_content = (
            f"Nome do Professor: {nome}.\n"
            f"Email de contato: {email}.\n"
            f"Áreas de Pesquisa e Interesse: {linha_pesquisa}.\n"
            f"Oferece bolsa de Iniciação Científica (IC)? {ic}.\n"
            f"Oferece bolsa de PDPD? {pdpd}."
        )

        metadata = {
            "nome": nome,
            "email": email,
            "pesquisa": linha_pesquisa,
            "row_id": index
        }

        doc = Document(page_content=page_content, metadata=metadata)
        documents.append(doc)

    # Vetorização
    print("Gerando Embeddings Otimizados...")
    embeddings_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    vector_store = FAISS.from_documents(documents, embeddings_model)

    print("✅ Base Vetorial Atualizada e Enriquecida!")

except Exception as e:
    print(f"❌ Erro no ETL: {e}")
```

```
# @title 4. Definição do Cérebro
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.documents import Document
from pydantic import BaseModel, Field
from thefuzz import process, fuzz
from typing import Optional

llm = ChatGroq(model="meta-llama/llama-4-maverick-17b-128e-instruct", temperature=0)

# ---> Agente Rewriter <---
rewriter_system = """
Sua tarefa é converter a entrada do usuário em uma busca direta no banco de dados.
REGRAS CRÍTICAS:
1. JAMAIS converse com o usuário. JAMAIS peça mais detalhes.
2. Se a entrada for "Conhece o X?", transforme em "Quais são as informações sobre o professor X?".
3. Se a entrada for curta ou vaga, assuma que é uma busca por professor ou tema.
4. Mantenha os nomes próprios EXATAMENTE como escritos.
5. Traduza as siglas, ou seja, transforme PLN em "Processamento de Linguagem Natural"
6. EM HIPOTESE NENHUMA diga algo mais além do que a pergunta reescrita.


Exemplos Obrigatórios:
Entrada: "Conhece o Donizeti?" -> Saída: Quais são as informações acadêmicas, e-mail e linha de pesquisa do professor Donizeti?
Entrada: "Quem é o prof Alexandre?" -> Saída: Quem é o professor Alexandre e qual sua área de atuação?
Entrada: "Sabe algo de redes?" -> Saída: Quais professores pesquisam sobre redes de computadores?
Entrada: "Opa, tudo bem?" -> Saída: Olá. (Se for apenas saudação, mantenha simples).

Entrada Atual: {question}
Histórico: {history}
Saída (Apenas o texto reescrito):
"""
rewriter_prompt = ChatPromptTemplate.from_template(rewriter_system)
rewriter_chain = rewriter_prompt | llm | StrOutputParser()

# ---> Agente Router <---
class IntentClassification(BaseModel):
    intent: str = Field(description="Categorias: 'FORA_DO_TEMA', 'BUSCA_PROFESSOR', 'BUSCA_TURMAS'")

router_prompt = ChatPromptTemplate.from_messages([
    ("system", "Classifique a intenção. Perguntas sobre pesquisa, orientação são BUSCA_PROFESSOR. Perguntas sobre turmas ou aulas do proximo quadrimestre, horarios de aula, matérias e ementas são BUSCA_TURMAS"),
    ("human", "{question}")
])
router_chain = router_prompt | llm.with_structured_output(IntentClassification)

# ---> Extrator de Entidade (NER) <---
class ProfessorEntity(BaseModel):
    nome: Optional[str] = Field(description="O nome do professor citado. Retorne None se não houver nome específico.")

extractor_system = """
Extraia APENAS o nome do professor citado.
Exemplos:
"Quais informações do professor Donizeti?" -> nome: "Donizeti"
"Me fale sobre o Alexandre Alves" -> nome: "Alexandre Alves"
"Quero alguém de IA" -> nome: None
"""
extractor_prompt = ChatPromptTemplate.from_messages([
    ("system", extractor_system),
    ("human", "{question}")
])
extractor_chain = extractor_prompt | llm.with_structured_output(ProfessorEntity)

# ---> Lógica Fuzzy <---
def buscar_por_nome_fuzzy(nome_extraido, dataframe, limit=5, threshold=60):
    """Busca focada no nome extraído."""
    if dataframe is None or 'Nome' not in dataframe.columns or not nome_extraido:
        return []

    nomes_existentes = dataframe['Nome'].dropna().unique().tolist()

    # Compara a string extraída com a lista de nomes reais da planilha
    resultados = process.extract(nome_extraido, nomes_existentes, scorer=fuzz.token_set_ratio, limit=limit)

    docs_fuzzy = []
    print(f"   🕵️ [DEBUG FUZZY] Buscando por: '{nome_extraido}'")

    for nome, score in resultados:
        if score >= threshold:
            print(f"       -> Match Fuzzy: '{nome}' (Score: {score}%)")
            row = dataframe[dataframe['Nome'] == nome].iloc[0]
            email = row.get('Email', row.get('Email ', 'Não informado'))

            conteudo = (
                f"**RESULTADO EXATO (ENCONTRADO POR NOME)**\n"
                f"Nome: {nome}\n"
                f"Email: {email}\n"
                f"Linha de Pesquisa: {row.get('Linha de Pesquisa', 'N/A')}\n"
                f"Status Bolsa IC: {row.get('Edital', 'Não informado')}\n"
                f"Status Bolsa PDPD: {row.get('Unnamed: 4', 'Não informado')}"
            )
            docs_fuzzy.append(Document(page_content=conteudo, metadata={"nome": nome, "origem": "fuzzy"}))

    return docs_fuzzy

# ---> Agente RAG <---
def recomendar_professor(query):

    # Extração de Entidade
    entity = extractor_chain.invoke(query)
    nome_identificado = entity.nome

    fuzzy_docs = []

    if nome_identificado:
        print(f"   🎯 [DEBUG NER] Nome identificado: '{nome_identificado}'.")
        if 'df' in globals():
            fuzzy_docs = buscar_por_nome_fuzzy(nome_identificado, df, limit=3, threshold=60)
    else:
        print(f"   ⚪ [DEBUG NER] Nenhum nome específico identificado. Pulando Fuzzy.")

    # Busca Vetorial
    print(f"   🔍 [DEBUG RAG] Buscando vetores...")
    retriever = vector_store.as_retriever(search_kwargs={"k": 5})
    vector_docs = retriever.invoke(query)

    final_docs = fuzzy_docs + vector_docs
    seen_names = set()
    unique_docs = []
    for d in final_docs:
        name = d.metadata.get("nome", "desconhecido")
        if name not in seen_names:
            unique_docs.append(d)
            seen_names.add(name)

    print(f"   📄 [DEBUG CONTEXTO] {len(unique_docs)} docs enviados.")

    context_text = "\n\n".join([f"--- Doc ---\n{d.page_content}" for d in unique_docs])

    rag_system = """
    Você é um assistente acadêmico da UFABC. Use o contexto para responder.

    1. Se houver um documento marcado como **RESULTADO EXATO**, use-o como fonte principal.
    2. Responda diretamente se o professor existe, qual sua linha de pesquisa e se oferece bolsas (IC/PDPD).

    Contexto:
    {context}
    """
    rag_prompt = ChatPromptTemplate.from_messages([("system", rag_system), ("human", "{question}")])
    rag_chain = rag_prompt | llm | StrOutputParser()
    return rag_chain.invoke({"context": context_text, "question": query})

# Placeholders
def responder_fora_do_tema(): return "Desculpe, meu foco é exclusivamente acadêmico."
def buscar_turmas(): return "Funcionalidade de busca de turmas em desenvolvimento."
```

```
# @title 5. Chat Interativo
import time

def main_loop():
    print("\n🎓 --- Assistente de Orientadores UFABC Iniciado (Modo Debug) ---")
    print("Digite 'sair' para encerrar.\n")

    chat_history = []

    while True:
        user_input = input("\n👤 Você: ")

        if user_input.lower() in ['sair', 'exit', 'tchau']:
            print("🤖 Assistente: Até logo! Boa sorte nas pesquisas.")
            break

        # Rewriter
        history_str = "\n".join([f"User: {h[0]}\nAI: {h[1]}" for h in chat_history[-3:]])

        print("   ⚙️ [Processando] Reformulando pergunta...", end="\r")
        rewritten_query = rewriter_chain.invoke({"history": history_str, "question": user_input})

        # DEBUG REWRITER
        print(f"\n   🔄 [DEBUG REWRITER] Pergunta Original: '{user_input}'")
        print(f"   🔄 [DEBUG REWRITER] Pergunta Reescrita: '{rewritten_query}'")

        # Router
        print("   ⚙️ [Processando] Classificando intenção...", end="\r")
        intent_result = router_chain.invoke({"question": rewritten_query})
        intent = intent_result.intent

        # DEBUG ROUTER
        print(f"   🔀 [DEBUG ROUTER] Intenção Detectada: {intent}")

        # Execução baseada na intenção
        response = ""

        if intent == "FORA_DO_TEMA":
            response = responder_fora_do_tema()

        elif intent == "BUSCA_TURMAS":
            response = perguntar(rewritten_query)

        elif intent == "BUSCA_PROFESSOR":
            response = recomendar_professor(rewritten_query)

        else:
            response = "Não consegui entender sua intenção. Poderia reformular?"

        # Exibe a resposta
        print(f"\n🤖 Assistente: {response}")
        print("-" * 50)

        chat_history.append((user_input, response))

if __name__ == "__main__":
    main_loop()
```
